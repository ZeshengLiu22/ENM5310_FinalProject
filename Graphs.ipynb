# Base CNN, Data Augmentation

```python
import jax
import jax.numpy as np
import numpy as onp
import matplotlib.pyplot as plt
```

```python
# load base CNN outs
base_CNN_loss = np.load('/content/base_CNN_log_loss.npy')
base_CNN_acc = np.load('/content/base_CNN_acc.npy')
base_CNN_test_loss = np.load('/content/base_CNN_test_log_loss.npy')
base_CNN_test_acc = np.load('/content/base_CNN_test_acc.npy')
```

```python
# load grayscale + base CNN outs
gs_loss = np.load('/content/grayscale_CNN_log_loss.npy')
gs_acc = np.load('/content/grayscale_CNN_acc.npy')
gs_test_loss = np.load('/content/grayscale_CNN_test_log_loss.npy')
gs_test_acc = np.load('/content/grayscale_CNN_test_acc.npy')
```

```python
# load brightness + base CNN outs
bt_loss = np.load('/content/bt_CNN_log_loss.npy')
bt_acc = np.load('/content/bt_CNN_acc.npy')
bt_test_loss = np.load('/content/bt_CNN_test_log_loss.npy')
bt_test_acc = np.load('/content/bt_CNN_test_acc.npy')
```

```python
# load mirroring + base CNN outs
mr_loss = np.load('/content/mr_CNN_log_loss.npy')
mr_acc = np.load('/content/mr_CNN_acc.npy')
mr_test_loss = np.load('/content/mr_CNN_test_log_loss.npy')
mr_test_acc = np.load('/content/mr_CNN_test_acc.npy')
```

```python
# base CNN + data augmentation plots
plt.figure(dpi = 150)
plt.plot(base_CNN_loss, label='CNN')
plt.plot(gs_loss, label='CNN + grayscale')
plt.plot(bt_loss, label='CNN + brightness')
plt.plot(mr_loss, label='CNN + mirroring')
plt.title('Training Losses')
plt.xlabel(r'Iterations ($\times 50$)')
plt.ylabel('Log Loss')
plt.legend()
```

```python
# base CNN + data augmentation plots
plt.figure(dpi = 150)
plt.plot(base_CNN_acc, label='CNN')
plt.plot(gs_acc, label='CNN + grayscale')
plt.plot(bt_acc, label='CNN + brightness')
plt.plot(mr_acc, label='CNN + mirroring')
plt.title('Training Accuracies')
plt.xlabel(r'Iterations ($\times 50$)')
plt.ylabel('Accuracy')
plt.legend()
```

```python
# base CNN + data augmentation plots
plt.figure(dpi = 150)
plt.plot(base_CNN_test_loss, label='CNN')
plt.plot(gs_test_loss, label='CNN + grayscale')
plt.plot(bt_test_loss, label='CNN + brightness')
plt.plot(mr_test_loss, label='CNN + mirroring')
plt.title('Testing Losses')
plt.xlabel(r'Iterations ($\times 50$)')
plt.ylabel('Log Loss')
plt.legend()
```

```python
# base CNN + data augmentation plots
plt.figure(dpi = 150)
plt.plot(base_CNN_test_acc, label='CNN')
plt.plot(gs_test_acc, label='CNN + grayscale')
plt.plot(bt_test_acc, label='CNN + brightness')
plt.plot(mr_test_acc, label='CNN + mirroring')
plt.title('Testing Accuracies')
plt.xlabel(r'Iterations ($\times 50$)')
plt.ylabel('Accuracy')
plt.legend()
```
